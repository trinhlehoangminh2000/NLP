**Text Cleaning**

1. Remove html tags
2. Remove Special Characters
3. Converting every thing to lower case
4. Removing Stop Words
5. Stemming



```python
def convert_lowercase(column):
    column = column.str.lower()
    return column
```

```python
tripadvisor_df['Review'][5]
```

```python
tripadvisor_df['Review'] = convert_lowercase(tripadvisor_df['Review'])
tripadvisor_df['Review'][5]
```

Removing HTML Tags

```python
def remove_html_tags(text):
    re_html = re.compile('<.*?>')
    return re_html.sub(r'', text)
```

```python
tripadvisor_df['Review'][5]
```

```python
tripadvisor_df['Review'] = tripadvisor_df['Review'].apply(remove_html_tags)
tripadvisor_df['Review'][5]
```

Remove Special Characters

```python
#Function to remove special characters
def remove_special(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
tripadvisor_df['Review'] = tripadvisor_df['Review'].apply(remove_special)
tripadvisor_df['Review'][1]
```

Spelling Correction 

```python
from textblob import TextBlob
tripadvisor_df['Review'] = TextBlob.correct(tripadvisor_df['Review'])
tripadvisor_df['Review'][2]
```
